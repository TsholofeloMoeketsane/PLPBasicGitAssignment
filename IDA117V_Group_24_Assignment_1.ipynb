**Feature Engineering & Transformation**

```
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import seaborn as sns
import pandas as pd


#Normalize/standardize the numerical features.
#Encode the categorical feature 'species' using one-hot encoding.

# Step 1: Load the Iris dataset
iris = sns.load_dataset('iris')

# Step 2: Normalize/Standardize the numerical features
numerical_features = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numerical_features)
scaled_features_df = pd.DataFrame(scaled_features, columns=numerical_features.columns)
print("Standardized Numerical Features:\n", scaled_features_df.head())

# Step 3: Encode the categorical feature 'species' using one-hot encoding
encoder = OneHotEncoder(sparse_output=False)
encoded_species = encoder.fit_transform(iris[['species']])
encoded_species_df = pd.DataFrame(encoded_species, columns=encoder.get_feature_names_out(['species']))
print("One-Hot Encoded Species:\n", encoded_species_df.head())

# Step 4: Combine the transformed features
transformed_iris = pd.concat([scaled_features_df, encoded_species_df], axis=1)
print("Transformed Iris Dataset:\n", transformed_iris.head())
```
